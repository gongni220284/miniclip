## Import

```python
import os
import json
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, Model,models
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint,ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

from sklearn.metrics import f1_score, classification_report, confusion_matrix
import seaborn as sns
from sklearn.manifold import TSNE


```

```python
# FONCTIONS UTILES

def preview_images_from_dir(image_dir, image_size=(224, 224), n=12,
                            subset="training", seed=123):
    """
    Affiche n images avec leurs labels à partir d'un dossier
    (flow_from_directory).
    N'utilise PAS train_gen pour ne pas en avancer l'index.
    """
    # Générateur léger juste pour l'aperçu
    # Ici il y a une bonne indication si vous regardez bien pour la classif :)
    preview_datagen = ImageDataGenerator(validation_split=0.2, rescale=1./255)

    preview_gen = preview_datagen.flow_from_directory(
        image_dir,
        target_size=image_size,
        batch_size=n,
        class_mode="categorical",
        subset=subset,
        shuffle=True,
        seed=seed
    )

    imgs, y = next(preview_gen)    # imgs: (n, H, W, 3), y: one-hot
    class_names = list(preview_gen.class_indices.keys())
    labels = np.argmax(y, axis=1)

    rows = 3
    cols = int(np.ceil(n / rows))
    plt.figure(figsize=(4*cols, 4*rows))
    for i in range(min(n, imgs.shape[0])):
        plt.subplot(rows, cols, i+1)
        plt.imshow(imgs[i])
        plt.title(class_names[labels[i]])
        plt.axis("off")
    plt.tight_layout()
    plt.show()
```

## import lien de image et label

```python
# Répertoire cible pour sauvegarder vos modèles
model_dir = "./models_forclip"
os.makedirs(model_dir, exist_ok=True)

DATA = "../data/augmented" 
preview_images_from_dir(DATA, image_size=(224,224), n=12, subset="training")

CLASSES = ["ball", "bike", "dog", "water"]

image_paths = []
labels = []

for label_id, cls in enumerate(CLASSES):
    cls_folder = os.path.join(DATA, cls)

    #["ball_001", "ball_002", ..., "ball_150"]["bike_001", "bike_002", ..., "bike_150"]...
    for item in os.listdir(cls_folder):
        item_path = os.path.join(cls_folder, item)
        if not os.path.isdir(item_path):
            continue
            
        for f in os.listdir(item_path):
            if f.endswith(".jpg"):
                image_paths.append(os.path.join(item_path, f))
                labels.append(label_id)
```

## classification & normaliser

```python
#stratify pour assurer chaque class ont les meme nombre de test
paths_train, paths_val, labels_train, labels_val = train_test_split(image_paths, labels, test_size=0.2, random_state=42, stratify=labels)

img_size = 224
batch_size = 32

def InputImg(path, label):
    img = tf.io.read_file(path)
    #R,G,B
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, (img_size, img_size))
    #normaliser
    img = img / 255.0
    return img, label
```

## CNN


```python

#Input
train_ds = tf.data.Dataset.from_tensor_slices((paths_train, labels_train))
#choisir le nombre de prefetch auto
train_ds = train_ds.shuffle(len(paths_train)).map(InputImg).batch(batch_size).prefetch(tf.data.AUTOTUNE)

val_ds = tf.data.Dataset.from_tensor_slices((paths_val, labels_val))
val_ds = val_ds.map(InputImg).batch(batch_size).prefetch(tf.data.AUTOTUNE)

```

```python
inputs = layers.Input(shape=(img_size, img_size, 3))


x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(inputs)
x = layers.BatchNormalization()(x)
x = layers.MaxPooling2D(2, 2)(x)
x = layers.Dropout(0.25)(x)


x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x)
x = layers.BatchNormalization()(x)
x = layers.MaxPooling2D(2, 2)(x)
x = layers.Dropout(0.25)(x)

x = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(x)
x = layers.BatchNormalization()(x)
x = layers.MaxPooling2D(2, 2)(x)
x = layers.Dropout(0.3)(x)


x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(128, activation='relu', 
                 kernel_regularizer=tf.keras.regularizers.l2(0.01))(x)
x = layers.Dropout(0.5)(x)

outputs = layers.Dense(len(CLASSES), activation='softmax')(x)

model = models.Model(inputs=inputs, outputs=outputs)
```

```python
model.summary()
```

```python
model.compile(
    optimizer=Adam(learning_rate=1e-4), 
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
#test
#history = model.fit(train_ds, epochs=1)
```

```python
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=8,
    restore_best_weights=True,
    verbose=1,
    min_delta=0.001
)

model_checkpoint = ModelCheckpoint(
    filepath=os.path.join(model_dir, "best_improved_cnn_model.keras"),
    monitor='val_accuracy',
    save_best_only=True,
    verbose=1
)
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.3,             
    patience=3,             
    min_lr=1e-7,
    verbose=1,
    cooldown=2           
)

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=50,
    callbacks=[early_stopping, model_checkpoint, reduce_lr],
    verbose=1
)
```

```python
# -------------------------------
# 学习曲线可视化
# -------------------------------

def plot_learning_curves(history):
    plt.figure(figsize=(14,5))

    # Loss
    plt.subplot(1, 2, 1)
    plt.plot(history.history["loss"], label="train_loss")
    plt.plot(history.history["val_loss"], label="val_loss")
    plt.title("Loss Curve")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.grid(True)

    # Accuracy
    plt.subplot(1, 2, 2)
    plt.plot(history.history["accuracy"], label="train_acc")
    plt.plot(history.history["val_accuracy"], label="val_acc")
    plt.title("Accuracy Curve")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

plot_learning_curves(history)
```

```python
# 检查模型在验证集上的准确率
val_ds_eval = tf.data.Dataset.from_tensor_slices((paths_val, labels_val))
val_ds_eval = val_ds_eval.map(InputImg).batch(32).prefetch(tf.data.AUTOTUNE)

y_pred = []
y_true = []

for imgs, labels in val_ds_eval:
    preds = model.predict(imgs, verbose=0)
    y_pred.extend(np.argmax(preds, axis=1))
    y_true.extend(labels.numpy())

from sklearn.metrics import accuracy_score, f1_score
acc = accuracy_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred, average='weighted')

print(f"准确率 (Accuracy): {acc:.4f}")
print(f"F1 分数: {f1:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import tensorflow as tf
from tensorflow.keras import Model

print("\n" + "="*80)
print("修复版：提取正确的特征层")
print("="*80 + "\n")

# ─────────────────────────────────────────────────────────────────────────
# 方法 1：使用层的名称直接指定（最安全）
# ─────────────────────────────────────────────────────────────────────────

print("方法：直接用 dense_1 层（256 维特征）")
print("-" * 80)

feature_extractor = Model(
    inputs=model.input,
    outputs=model.get_layer('dense_1').output  # ← 直接用名字获取
)

print(f"✅ 特征提取器创建成功")
print(f"   输入形状: {feature_extractor.input_shape}")
print(f"   输出形状: {feature_extractor.output_shape}")
print(f"   提取的层: dense_1 (256 维)")

print("\n" + "="*80)
print("步骤 1：重新创建验证集")
print("="*80 + "\n")

val_ds_tsne = tf.data.Dataset.from_tensor_slices((paths_val, labels_val))
val_ds_tsne = val_ds_tsne.map(InputImg).batch(32).prefetch(tf.data.AUTOTUNE)

print(f"✅ 验证集创建完成 ({len(paths_val)} 个样本)")

print("\n" + "="*80)
print("步骤 2：提取特征")
print("="*80 + "\n")

features = []
y_true_list = []

for batch_idx, (imgs, labels) in enumerate(val_ds_tsne):
    feats = feature_extractor.predict(imgs, verbose=0)
    features.append(feats)
    y_true_list.extend(labels.numpy())
    
    if (batch_idx + 1) % 3 == 0:
        print(f"  已处理 {batch_idx + 1} 个 batch，样本数: {len(y_true_list)}")

features = np.vstack(features)
y_true_list = np.array(y_true_list)

print(f"\n✅ 特征提取完成！")
print(f"   特征形状: {features.shape}")
print(f"   标签形状: {y_true_list.shape}")
print(f"   特征维度: {features.shape[1]} ← 这应该是 256")
print(f"\n   标签分布:")
for i in range(len(CLASSES)):
    count = np.sum(y_true_list == i)
    print(f"     - {CLASSES[i]}: {count}")

print("\n" + "="*80)
print("步骤 3：运行 t-SNE")
print("="*80 + "\n")

n_samples = features.shape[0]
perplexity = min(30, (n_samples - 1) / 3)

print(f"t-SNE 参数：")
print(f"   样本数: {n_samples}")
print(f"   特征维度: {features.shape[1]} ✅（256 维）")
print(f"   Perplexity: {perplexity:.1f}")
print(f"\n运行中... (需要 30 秒到 2 分钟)")

tsne = TSNE(
    n_components=2,
    perplexity=perplexity,
    random_state=42,
    max_iter=1000,
    n_jobs=-1,
    verbose=1
)

features_2d = tsne.fit_transform(features)

print(f"\n✅ t-SNE 完成！")
print(f"   输出形状: {features_2d.shape}")

print("\n" + "="*80)
print("步骤 4：绘制结果")
print("="*80 + "\n")

fig, ax = plt.subplots(figsize=(15, 13))

colors = ['red', 'green', 'blue', 'yellow']  # 更漂亮的颜色

for i, class_name in enumerate(CLASSES):
    mask = y_true_list == i
    count = np.sum(mask)
    
    print(f"绘制 {class_name}: {count} 个点...")
    
    if count > 0:
        ax.scatter(
            features_2d[mask, 0],
            features_2d[mask, 1],
            label=f"{class_name} (n={count})",
            s=150,  # 点的大小
            alpha=0.75,
            c=colors[i],
            edgecolors='black',
            linewidth=1.2,
            rasterized=True
        )

ax.set_xlabel('t-SNE Dimension 1', fontsize=14, fontweight='bold')
ax.set_ylabel('t-SNE Dimension 2', fontsize=14, fontweight='bold')
ax.set_title(
    f't-SNE Visualization - 256D Features (dense_1 layer)',
    fontsize=16,
    fontweight='bold',
    pad=20
)
ax.legend(fontsize=13, loc='best', framealpha=0.98, edgecolor='black')
ax.grid(True, alpha=0.3, linestyle='--')

plt.tight_layout()

# 保存图片
output_path = os.path.join(model_dir, 'tsne_visualization_fixed.png')
plt.savefig(output_path, dpi=150, bbox_inches='tight')
print(f"\n✅ 图片已保存到: {output_path}")

plt.show()

print("\n" + "="*80)
print("✅ 完成！现在应该能看到四个清晰分开的点云了")
print("="*80)
```

```python
from PIL import Image
def predict_image(image_path):
    img = Image.open(image_path).convert("RGB")
    img = img.resize((img_size, img_size))
    img = np.array(img, dtype=np.float32) / 255.0
    img = np.expand_dims(img, axis=0)

    preds = model.predict(img)[0]
    class_id = np.argmax(preds)
    confidence = preds[class_id]

    print("class：", CLASSES[class_id])
    print("confidence：", confidence)
    return CLASSES[class_id], confidence
```
